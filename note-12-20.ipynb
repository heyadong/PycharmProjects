```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
fig,ax = plt.subplots(figsize=(7,7))
widths = df['花萼宽度'][:50]
lengths = df['花萼长度'][:50]
ax.scatter(widths, lengths)
ax.set_xlabel('花萼宽度')
ax.set_ylabel('花萼长度')
```

```python
import statsmodels.api as sm
```

```python
y = lengths
x = widths
x = sm.add_constant(x)
results = sm.OLS(y,x).fit()
print(results.summary())
```
